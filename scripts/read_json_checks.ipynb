```python
def fix_col_titles(df: pd.DataFrame) -> pd.DataFrame:
    dest_cols = []
    for col in df.columns:
        new_name = (
            col.lower()
            .replace("-", "_")
            .replace(" ", "_")
            .replace("___", "_")
            .replace("(", "")
            .replace(")", "")
            # new ones
            .replace("'", "")
            .replace("__", "_")
        )
        dest_cols.append(new_name)
        df[new_name] = df[col]

    return df[dest_cols]
```

```python
new_df = fix_col_titles(df)
```
