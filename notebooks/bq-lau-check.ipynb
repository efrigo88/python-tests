```python
from google.cloud import bigquery
from google.oauth2 import service_account


def query_bigquery(query, target_project):

    path_to_credentials_json = (
        "/Users/emilianofrigo/heroes-analytics-dev-260a66cca2f5.json"
    )

    credentials = service_account.Credentials.from_service_account_file(
        path_to_credentials_json
    )

    client = bigquery.Client(credentials=credentials, project=target_project)

    try:
        query_job = client.query(query)

        result = query_job.result()

        rows = [dict(row) for row in result]
        return rows

    except Exception as e:
        print(f"An error occurred: {e}")
        return None


query = "SELECT * FROM `heroes-analytics-prod.external_sources.seller_central__get_sales_and_traffic_report` LIMIT 10"
target_project = "heroes-analytics-prod"

result = query_bigquery(query, target_project)

if result:
    print("Query Results:")
    for row in result:
        print(row)
else:
    print("Failed to retrieve results from BigQuery.")
```
