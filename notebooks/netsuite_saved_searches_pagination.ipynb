```python
import requests
import json
from requests_oauthlib import OAuth1Session
```

## Configuration

```python
"""
Sample GET request:

    GET Request
    curl --location 'https://6914496.restlets.api.netsuite.com/app/site/hosting/restlet.nl?script=1818&deploy=1&searchID=8752&rowNum=1000&pageNum=1' \
    --header 'Content-Type: application/json' \
    --header 'Cookie: NS_ROUTING_VERSION=LAGGING' \
    --header 'Authorization: Same OAuth 
"""
```

```python
# Replace these values with your actual OAuth credentials
consumer_key = ''
consumer_secret = ''
token_id = ''
token_secret = ''
account_id = '6914496'

# Create an OAuth1Session instance with your credentials
oauth = OAuth1Session(
    client_key=consumer_key,
    client_secret=consumer_secret,
    resource_owner_key=token_id,
    resource_owner_secret=token_secret,
    realm=account_id,
    signature_method="HMAC-SHA256",
)
```

```python
# Config URL params
script = "1818"
deploy = "1"
search_id = "8752"
row_num = "1000"
page_num = "1"

# Full URL
base_url = "https://6914496.restlets.api.netsuite.com/app/site/hosting"
params_url = f"restlet.nl?script={script}&deploy={deploy}&searchID={search_id}&rowNum={row_num}&pageNum={page_num}"
```

```python
# Get the report
url = f"{base_url}/{params_url}"

# Make a GET request using the OAuth1Session instance
response = oauth.get(url)

# Check the response status
if response.status_code == 200:
    # Process the response data
    data = response.json()
    print(type(data))
    print()
    print(data)
else:
    print(f"Error: {response.status_code} - {response.text}")
```

```python
def count_results_elements(json_data):
    try:
        # Check if "results" key exists and is a list
        if "results" in json_data and isinstance(json_data["results"], list):
            # Count the number of elements in the "results" array
            results_count = len(json_data["results"])
            return results_count
        else:
            print("Invalid JSON structure. 'results' key not found or not a list.")
            return None

    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```

```python
count_results = count_results_elements(data)
print(count_results)
```
