```python
import re
```

```python
def remove_non_alphanumeric(column_name: str) -> str:
        """Removes all special characters except "_".

        Args:
            column_name (str): The string column we want to clean.

        Returns:
            str: Cleaned string column.
        """
        return re.sub(r"[^a-zA-Z0-9_]+", "", column_name)
```

```python
def clean_column_name(column_name: str) -> str:
            replacements = [
                (" ", "_"),
                ("[", "_"),
                ("]", "_"),
                ("(", "_"),
                (")", "_"),
                ("{", "_"),
                ("}", "_"),
                (":", "_"),
                (";", "_"),
                ("_-_", "_"),
                ("/", "_"),
            ]
            for old, new in replacements:
                column_name = column_name.replace(old, new)
                column_name = remove_non_alphanumeric(column_name)
            return column_name
```
