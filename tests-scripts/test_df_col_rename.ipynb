```python
from datetime import datetime

def spapi_df_transform(
    df: pd.DataFrame,
    brand_id: str,
    region_id: str,
    aud_process_ts: datetime = datetime.now(),
) -> pd.DataFrame:
    # Column standardization for Restock Inventory Report
    df.rename(columns = {"Country/Region Code": "country_region_code"}, inplace = True)

    df["brand_id"] = brand_id.upper()
    df["region_id"] = region_id.upper()
    df["aud_process_ts"] = aud_process_ts.strftime(
        "%Y-%m-%d %H:%M:%S"
    )

    return df
```

```python
import pytz

df_output = spapi_df_transform(df_test, "BARVIVO", "US", datetime(2021, 9, 23, 1, 55, 19, tzinfo=pytz.utc))
```
